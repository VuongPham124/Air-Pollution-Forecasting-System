```python
import pandas as pd

df = pd.read_csv("/home/tee/doanbigdata/Air_quality_official.csv")

columns_to_fix = ["CO(GT)", "PT08.S1(CO)", "NMHC(GT)", "C6H6(GT)", "PT08.S2(NMHC)", 
                  "NOx(GT)", "PT08.S3(NOx)", "NO2(GT)", "PT08.S4(NO2)", "PT08.S5(O3)", "T", "RH", "AH"]

# Thay thế giá trị âm bằng NaN
df[columns_to_fix] = df[columns_to_fix].mask(df[columns_to_fix] <= 0, None)

# Thay thế các giá trị NaN bằng giá trị trung bình của từng cột
df[columns_to_fix] = df[columns_to_fix].fillna(df[columns_to_fix].mean())

# Làm tròn về 1 số sau dấu phẩy
df[columns_to_fix] = df[columns_to_fix].round(1)

# Lưu lại dữ liệu đã xử lý
df.to_csv("/home/tee/doanbigdata/Air_quality_final.csv", index=False, header=True)
print(df.head(100))
```
